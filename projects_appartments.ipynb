3.5. –î–æ–±–∞–≤–∏—Ç—å —Å—Ç–æ–ª–±–µ—Ü —Å —Ç–∏–ø–æ–º —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã ('–ø–µ—Ä–≤—ã–π', '–ø–æ—Å–ª–µ–¥–Ω–∏–π', '–¥—Ä—É–≥–æ–π')

```python
#–Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö
def categorize_floor(row):
    try:
        floor = row[8]
        floors_total = row[6]
        if floor == 1:
            return '–ø–µ—Ä–≤—ã–π'
        elif floor == floors_total:
            return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
        return '–¥—Ä—É–≥–æ–π'
    except:
        pass
```

```python
data['floor_category'] = data.apply(categorize_floor, axis=1)
```
